```python
import numpy as np

def f(t):
    return 70 * np.exp(-1.5 * t) + 2.2 * np.exp(-0.075 * t) - 9

def bisection(func, a, b, tol):
    if func(a) * func(b) >= 0:
        print("Intervalo inválido!")
        return None

    while (b - a) >= tol:
        c = (a + b) / 2

        if func(c) == 0.0:
            return c

        if func(c) * func(a) < 0:
            b = c
        else:
            a = c

    return (a + b) / 2


intervalo_a = 1.52
intervalo_b = 1.54
tolerancia = 0.0005

raiz = bisection(f, intervalo_a, intervalo_b, tolerancia)

if raiz is not None:
    print(f"A raiz é aproximadamente: {raiz:.6f}")
else:
    print("Não foi possível encontrar a raiz dentro da tolerância especificada.")
```
